```python
import numpy as np
import matplotlib.pyplot as plt

n_satellites = 10
data_size = 100
phi = 0.7

linear = []
quadratic = []
quadratic_corrected = []
n_values = []

final_data_size = data_size * phi

step_linear = (data_size - final_data_size) / n_satellites
step_quadratic = np.power(phi, 1/n_satellites)

data_size_linear = data_size
data_size_quadratic = data_size

print(f"data_size_linear data_size_quadratic")

n_values.append(0)
linear.append(data_size_linear)
quadratic.append(data_size_quadratic)
print(f"{data_size_linear} {data_size_quadratic}")

for satellite in range(1, n_satellites + 1, 1):
    data_size_linear = data_size_linear - step_linear
    data_size_quadratic = data_size_quadratic * step_quadratic

    linear.append(data_size_linear)
    quadratic.append(data_size_quadratic)
    n_values.append(satellite)

    print(f"{data_size_linear} {data_size_quadratic}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, linear, 'b-o', label='Data Size Even (Linear)', markersize=4)
plt.plot(n_values, quadratic, 'r-s', label='Phi Even (Quadratic)', markersize=4)
plt.xlabel('Number of Satellites')
plt.ylabel('Data Size')
plt.title('Linear vs Quadratic Data Size Reduction')
plt.legend()
plt.grid(True, alpha=0.3)
```
