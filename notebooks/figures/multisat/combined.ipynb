```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from all three scripts
df1 = pd.read_csv('script1_results.csv')
df2 = pd.read_csv('script2_results.csv')
df3 = pd.read_csv('script3_results.csv')

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Time Costs
ax1.plot(df1['n_satellites'], df1['time_cost'], marker='o', label='Even Offloading', linewidth=2)
ax1.plot(df2['n_satellites'], df2['time_cost'], marker='s', label='First Offloading', linewidth=2)
ax1.plot(df3['n_satellites'], df3['time_cost'], marker='^', label='Last Offloading', linewidth=2)
ax1.set_xlabel('Number of Satellites', fontsize=12)
ax1.set_ylabel('Time Cost', fontsize=12)
ax1.set_title('Time Cost Comparison', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Energy Costs
ax2.plot(df1['n_satellites'], df1['energy_cost'], marker='o', label='Even Offloading', linewidth=2)
ax2.plot(df2['n_satellites'], df2['energy_cost'], marker='s', label='First Offloading', linewidth=2)
ax2.plot(df3['n_satellites'], df3['energy_cost'], marker='^', label='Last Offloading', linewidth=2)
ax2.set_xlabel('Number of Satellites', fontsize=12)
ax2.set_ylabel('Energy Cost', fontsize=12)
ax2.set_title('Energy Cost Comparison', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('combined_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("Combined plot saved as 'combined_results.png'")

# Optional: Print summary statistics
print("\n=== Summary Statistics ===")
print("\nTime Costs:")
print(f"Even Offloading: Min={df1['time_cost'].min():.2e}, Max={df1['time_cost'].max():.2e}")
print(f"First Offloading: Min={df2['time_cost'].min():.2e}, Max={df2['time_cost'].max():.2e}")
print(f"Last Offloading: Min={df3['time_cost'].min():.2e}, Max={df3['time_cost'].max():.2e}")

print("\nEnergy Costs:")
print(f"Even Offloading: Min={df1['energy_cost'].min():.2e}, Max={df1['energy_cost'].max():.2e}")
print(f"First Offloading: Min={df2['energy_cost'].min():.2e}, Max={df2['energy_cost'].max():.2e}")
print(f"Last Offloading: Min={df3['energy_cost'].min():.2e}, Max={df3['energy_cost'].max():.2e}")
```
