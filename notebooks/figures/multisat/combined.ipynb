```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from all three scripts
df1 = pd.read_csv('first_results.csv')
df2 = pd.read_csv('last_results.csv')
df3 = pd.read_csv('even_quadratic_results.csv')
df4 = pd.read_csv('even_linear_results.csv')

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Time Costs
ax1.plot(df1['n_satellites'], df1['time_cost'], marker='o', label='First Offloading', linewidth=2)
ax1.plot(df2['n_satellites'], df2['time_cost'], marker='s', label='Last Offloading', linewidth=2)
ax1.plot(df3['n_satellites'], df3['time_cost'], marker='^', label='Even Quadratic Offloading', linewidth=2)
ax1.plot(df4['n_satellites'], df4['time_cost'], marker='*', label='Even Linear Offloading', linewidth=2)
ax1.set_xlabel('Number of Satellites', fontsize=12)
ax1.set_ylabel('Time Cost (s)', fontsize=12)
ax1.set_title('Time Cost Comparison', fontsize=14, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Energy Costs (changed from ax1 to ax2)
ax2.plot(df1['n_satellites'], df1['energy_cost'], marker='o', label='First Offloading', linewidth=2)
ax2.plot(df2['n_satellites'], df2['energy_cost'], marker='s', label='Last Offloading', linewidth=2)
ax2.plot(df3['n_satellites'], df3['energy_cost'], marker='^', label='Even Quadratic Offloading', linewidth=2)
ax2.plot(df4['n_satellites'], df4['energy_cost'], marker='*', label='Even Linear Offloading', linewidth=2)
ax2.set_xlabel('Number of Satellites', fontsize=12)
ax2.set_ylabel('Energy Cost (J)', fontsize=12)
ax2.set_title('Energy Cost Comparison', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('combined_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
